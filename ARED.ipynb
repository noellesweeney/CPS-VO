```python
import matplotlib.pyplot as plt
import csv
import re
#import numpy as np
 
x = []
y = []
 
#data1 = np.genfromtxt('/Users/noell/OneDrive/Documents/Research/ARED/expE.csv')
#print(data1)

filePath = '/Users/noell/OneDrive/Documents/Research/ARED/expE.csv'
expID = re.search('exp([A-Z])\.csv', filePath).group(1)

from pandas import read_csv
df = read_csv(filePath)
#data = df.values
#print(data)

#Trajectory of All Vehicles
f = plt.figure()
f.set_figwidth(20)

plt.plot(df['timestamp'], df['cam_distance']%262, color = 'lightgray', linestyle = 'dashed', linewidth = '.001',
         marker = ',')

#CAT Vehicle
df.set_index("model", inplace=True)
cat = df.loc['Escape_Hybrid']
plt.plot(cat['timestamp'], cat['cam_distance']%262, color = 'red', linestyle = 'dashed', linewidth = '.001',
         marker = ',')

plt.xticks(rotation = 25)
plt.xlabel('Time')
plt.ylabel('Position on Road')
plt.title('(a) Trajectory of All Vehicles in Experiment ' + expID, fontsize = 20)

plt.show()


#Velocity Profile
f = plt.figure()
f.set_figwidth(20)
 
plt.plot(df['timestamp'], abs(df['cam_velocity']), color = 'lightgray', linestyle = 'dashed', linewidth = '.001',
         marker = ',')
#CAT Vehicle
plt.plot(cat['timestamp'], abs(cat['cam_velocity']), color = 'red', linestyle = 'dashed', linewidth = '.001',
         marker = ',')

plt.xticks(rotation = 25)
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('(b) Velocity Profile of All Vehicles in Experiment ' + expID, fontsize = 20)

plt.show()
```
